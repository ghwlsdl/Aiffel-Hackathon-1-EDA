```python
young = ((PatientInfo.age=='20s') | (PatientInfo.age=='30s'))& (PatientInfo.province=='Seoul')
```

```python
PatientInfo[young]
```

```python
grouped = PatientInfo[young].groupby(PatientInfo['city'])
grouped
```
